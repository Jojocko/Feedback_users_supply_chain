# load labelled dataset

```python
import pandas as pd
!pip install itables
from itables import show
import ast
import numpy as np

df = pd.read_csv('../data/chatgpt_labelled_reviews_100/train.csv')
df_test = pd.read_csv('../data/chatgpt_labelled_reviews_100/test.csv')
#df = df.drop(["prompt tokens","completion tokens", "hallucinated labels count", "human checked ?", "id", "star"],axis=1)
#df = df.dropna(axis=0)
#df['chatgpt_label'] = df['chatgpt_label'].apply(lambda x: ast.literal_eval(x)[0] if len(ast.literal_eval(x))>0 else np.nan)
df = df.drop(["Unnamed: 0"],axis=1)
df_test = df_test.drop(["Unnamed: 0"],axis=1)

display(df.sample(5))
print('train shape: ', df.shape)
print('test shape: ', df_test.shape)
#df.info()
#display(df['chatgpt_label'].value_counts(normalize=True))
#display(df['chatgpt_label'].value_counts())
#print('nb of review without label:', df['chatgpt_label'].isna().sum())

#df[df['chatgpt_label'].isna()] = '<no label>'

import seaborn as sns
sns.displot(y='chatgpt_label', data=df);
sns.displot(y='chatgpt_label', data=df_test);

# TODO processs label
```

# Classification avec réseau dense après fine-tuning

```python
!pip install sentence-transformers
from sentence_transformers import SentenceTransformer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler

model_finetuned = SentenceTransformer('../models/fine-tuned - train_len=160,  pairs_sample_size=1300 - epochs=20, batch_size=32, base_model=distilbert-multilingual-nli-stsb-quora-ranking')

# X = model_finetuned.encode(df['review'].tolist(), convert_to_tensor=False)

# label_encoder = LabelEncoder()
# y_encoded = label_encoder.fit_transform(df['chatgpt_label'])

# X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

X_train = model_finetuned.encode(df['review'].to_numpy(), convert_to_tensor=False)
X_test = model_finetuned.encode(df_test['review'].to_numpy(), convert_to_tensor=False)

label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(df['chatgpt_label'])
y_test = label_encoder.transform(df_test['chatgpt_label'])

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.metrics import Recall, Precision
import tensorflow as tf

model_nn = Sequential()
model_nn.add(Dense(32, activation='relu', input_shape=(X_train_scaled.shape[1],)))
model_nn.add(Dropout(0.3))
model_nn.add(Dense(len(label_encoder.classes_), activation='softmax'))

model_nn.compile(optimizer='adam',
                 loss='sparse_categorical_crossentropy',
                 metrics=['accuracy'])

early_stopping = EarlyStopping(monitor='val_loss', patience=3, verbose=1, mode='min', restore_best_weights=True)

history = model_nn.fit(X_train_scaled, y_train, epochs=30, batch_size=32, validation_split=0.2, callbacks=[early_stopping])
```

Le modèle présente une précision de validation de 84%, ce qui indique une amélioration par rapport aux versions précédentes du DNN sans fine tuning. Cela pourrait suggérer que le modèle a la capacité de bien généraliser.

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('History')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```

```python
plt.plot(history.history['accuracy'], label='Training acc')
plt.plot(history.history['val_accuracy'], label='Validation acc')
plt.title('History')
plt.ylabel('acc')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```

```python
# Évaluation sur l'ensemble de test

test_loss, test_accuracy = model_nn.evaluate(X_test_scaled, y_test)
print("Test Accuracy:", test_accuracy)
```

La précision sur l'ensemble de test étant de seulement 40%, il est clair que le modèle ne performe pas bien sur de nouvelles données.
